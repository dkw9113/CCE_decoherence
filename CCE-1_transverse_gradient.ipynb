# Legend:

 $B_0$ is a constant magnetic field applied along z (quantization axis of NV)
 
 $\Delta B$ is a magnetic field gradient
 
 Plots show the decoherence shape when including non-zero transverse Zeeman splitting (due to gradient) - BLUE CURVES and when including gradient only as an addition to $\omega_z$ (green curve). 
 
 Second plot in each panel shows $W_1^{incl}-W_1^{not~incl.}$ i.e. difference of coherence curves for the case when we include the transverse component of Zeeman splitting and when we are not (Described as: CCE-1: Differences when including $\omega_x\neq 0$)


# $B_0 = 0.3 \text{T}$, $\Delta B = 6\cdot 10^{-4} \frac{\text{T}}{\text{nm}}$ 

```python
import numpy as np
import matplotlib.pyplot as plt
included=np.loadtxt("transverse_included.dat")
not_included=np.loadtxt("transverse_not_included.dat")
%matplotlib inline
```

```python
from IPython.zmq.pylab import backend_inline
cfg=backend_inline.InlineBackendConfig.instance()
cfg.rc.update({'savefig.dpi':144})
plot
```

```python
plt.figure(figsize=(10,6))
plt.plot(included[:,0],included[:,1],'b',label=r"$\omega_x\neq0$")
plt.plot(not_included[:,0],not_included[:,1],'g',label=r"$\omega_x=0$")
plt.legend()
```

```python
plt.figure(figsize=(10,6))
plt.plot(included[:,0],included[:,1]-not_included[:,1])
plt.title(r"CCE-1: Differences when including $\omega_x\neq0$")
```
