# $B_0 = 0.3 \text{T}$, $\Delta B = 6\cdot 10^{-4} \frac{\text{T}}{\text{nm}}$ 

```python
import numpy as np
import matplotlib.pyplot as plt
included=np.loadtxt("transverse_included.dat")
not_included=np.loadtxt("transverse_not_included.dat")
%matplotlib inline
```

```python
plt.plot(included[:,0],included[:,1],'b',label=r"$\omega_x\neq0$")
plt.plot(not_included[:,0],not_included[:,1],'g',label=r"$\omega_x=0$")
plt.legend()
```

```python
plt.plot(included[:,0],included[:,1]-not_included[:,1])
plt.title(r"CCE-1: Differences when including $\omega_x\neq0$")
```
