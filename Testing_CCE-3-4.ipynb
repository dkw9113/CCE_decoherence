```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
a="cce-3_noaniso_result.dat"
b=pd.read_csv(a,delimiter="\t",header=None)
time=np.loadtxt("Wan_result_PY.dat")[:,0]
for i in range(3):
    for j in range(len(b.loc[:,i])):
        b.set_value(j,i,eval(b.loc[j,i]))
for i in range(3):
    for j in range(len(b.ix[:,i])):
        b.set_value(j,i,abs(complex(b.ix[j,i][0],b.ix[j,i][1])))
    #print(b.ix[:,i])
plt.title("B= 3000 G. Aniso not included.")
plt.plot(time,b.ix[:,0],color="green",label="CCE-1")
plt.plot(time,b.ix[:,1],color="red",label="CCE-2")
plt.plot(time,b.ix[:,2],color="blue",ls="--",label="CCE-3")
#plt.plot(time,b.ix[:,3],color="magenta",label="CCE-2 anal")
plt.ylim(-0.1,1.1)
plt.legend()
#plt.show()
```

```python
aniso=np.loadtxt("cce-3_aniso_result.dat")
plt.title("B = 100 G. Aniso included.")
plt.plot(aniso[:,0],aniso[:,1],color="green",label="CCE-1")

for i in range(len(aniso[:,3])):
    if aniso[i,2]>aniso[i,4]:
        aniso[i,2]=aniso[i,4]
plt.plot(aniso[:,0],aniso[:,2],color="magenta",label="CCE-2")
for i in range(len(aniso[:,3])):
    if aniso[i,3]>aniso[i,4]:
        aniso[i,3]=aniso[i,4]
plt.plot(aniso[:,0],aniso[:,3],color="cyan",ls="--",label="CCE-3")
plt.plot(aniso[:,0],aniso[:,4],color="blue",ls="--",label="CCE-2 anal")
plt.ylim(-0.1,1.1)
plt.legend()
```
